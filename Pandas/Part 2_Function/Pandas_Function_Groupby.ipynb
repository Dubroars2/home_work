```python
#Задание 1
"""
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class
"""
```

```python
import pandas as pd
```

```python
data_rating = pd.read_csv("ratings.csv")
data_rating.head(2)
```

```python
def class_movies(check_list):
    if check_list["rating"] <= 2:
        return "низкий рейтинг "
    if check_list["rating"] <= 4 and check_list["rating"] > 2 :
        return "средний рейтинг "
    if check_list["rating"] >= 4.5 :
        return "высокий рейтинг "
```

```python
data_rating["class"] = data_rating.apply(class_movies, axis = 1)
data_rating.head()
```

```python
# Задание 2
"""
Используем файл keywords.csv.
Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. 
Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. 
Если поисковый запрос не содержит названия города, то ставим ‘undefined’.
Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.
"""
```

```python
df_reg = pd.read_csv("keywords.csv")
df_reg.tail(10)
```

```python
def reg_function(col):
    if ("москва" in col["keyword"]) or ("тула" in col["keyword"]) or ("ярославль" in col["keyword"]):
        return "Центр" 
    
    if ("петербург" in col["keyword"]) or ("псков" in col["keyword"]) or ("мурманск" in col["keyword"]):
        return "Северо-Запад" 
  
    if ("владивосток" in col["keyword"]) or ("сахалин" in col["keyword"]) or ("хабаровск" in col["keyword"]):
        return "Дальний Восток"    
    return "undefined"
```

```python
df_reg["Region"] = df_reg.apply(reg_function, axis = 1)
df_reg.head()
```
